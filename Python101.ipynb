```python
#First you need to set the parameters of the gambling game

#In this game, when you win you gain a dollar
#when you loose you loose a dollar

#The probability of winning is determined by
#the variable named prob_win
#in this simulation you win with a
#probabilityof 30%
prob_win=.5

#The starting amount of money is set by the variable
#called start_money
start_money=500

#You also need away to keep track of the
#Current amount of money the user has throughout the simulation
#We’ll call this curr_money and initially it should
#be the same value as the starting money
curr_money = start_money

#The amount of money the player wants to get before stopping
goal=1000

#The simulation should run until either the player reached
#their goal or runs out of money

#These, therefore should be the conditions for the while loop
while curr_money > 0 and curr_money < goal:
    #first get a random float from0to1
    
    random_draw = random.random()
    #based on the probability of winning specified above
    #see if the random draw resulted in a win or loss for the
    #player
    if random_draw < prob_win:
        curr_money = curr_money + 1
    else:
        curr_money = curr_money - 1
    #After the while loop finishes, tell the player if they won or lost
if curr_money == 0:
    print("Sorry, you lost all of your money")
else:
    print("Congratulations! You won "+str(curr_money - start_money)+" dollars.")
```

Lets run a lot of gambling simulations

```python
# define how many rounds of simulations to run
num_rounds = 1000

#The amount of money the player wants to get before stopping
goal=550

#probability of winning still 30%
prob_win=.48

#starting money for each round
start_money=500

#total winnings
#we will track this during all our simulations, starting with 0
total_money = 0

#we should also keep track of how many games we have simulated, starting with 0
total_games_played = 0 

#we should also track the number games we come out ahead. starting from 0
total_wins = 0

for sim_num in range(num_rounds):
    
    # start with the same amount each time
    curr_money = start_money

    # track the number of games playe dthis round
    games_played = 0
    
    # play until we're out of money
    while curr_money > 0 and curr_money < goal:
        #first get a random float from0to1
        random_draw = random.random()
        
        #based on the probability of winning specified above
        #see if the random draw resulted in a win or loss for the
        #player
        if random_draw < prob_win:
            curr_money = curr_money + 1
        else:
            curr_money = curr_money - 1
        
        # update the number of games played
        games_played = games_played + 1
        
    #After the while loop finishes if we won
    if curr_money != 0:
        # record the win
        total_wins = total_wins + 1
        
    # record the number of games played in the simulation
    total_games_played = total_games_played + games_played

    # record the total winnings
    total_money = total_money + curr_money

prob_win = total_wins / num_rounds

average_money = total_money / num_rounds

average_games = total_games_played / num_rounds

print('A player has a ', prob_win, 'chance of hitting their goal of ',
      goal, ' dollars if they start with ', start_money, ' dollars.')

print('They play ', average_games, ' games each round on average.')

print('They win ', average_money, ' on average per round.')
```

```python
def diagonalize(X, n_dims):
    X = X - np.mean(X) # I've actually already done this to the example data but you can do it again
    cov = np.cov(X.T)
    eigvals, eigvecs = np.linalg.eig(cov)
    sort_order = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sort_order]
    cumvar = np.cumsum(eigvals/np.sum(eigvals))
    eigvecs = eigvecs[:, sort_order]
    eigen_modes = X.dot(eigvecs[:, :n_dims])
    
    return cumvar[:n_dims], eigvecs[:, :n_dims], eigen_modes, cov
    
    
```
